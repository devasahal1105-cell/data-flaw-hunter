```python
print("Missing values per column:")
print(df.isnull().sum())
print("\nMissing percentage:")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
print("Total rows:", len(df))
print("Total missing values:", df.isnull().sum().sum())
print("Overall missing %:", round(df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100, 2))
print("\nUnique cities:", df['City'].nunique())
print("Cities:", df['City'].unique())
print("\nDate range:", df['Date'].min(), "to", df['Date'].max())
```

## What I Found on Day 2

**Dataset:** India Air Quality | 26 cities | 2015–2020

### Real Flaws Spotted:
- 88,488 missing values — 18.73% of entire dataset
- Xylene missing 61% of readings
- PM10 missing 37% — a critical pollution metric
- AQI max value = 2049 — impossible (scale only goes to 500)
- PM10 max = 1000 — suspiciously round, likely a sensor cap error
- NO and O3 min = 0.00 — near impossible in Indian city air
